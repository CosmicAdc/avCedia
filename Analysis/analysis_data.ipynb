Partición del dataset 

```python
import pandas as pd
import numpy as np

# Lee el archivo CSV
df = pd.read_csv("/home/ubuntu/CEDIA/avcedia/streamlit_project/data_process/QA_FINETUNNING_AV_pt1.csv", sep=";")

# Calcula los puntos de división para 100 partes
split_points = [i * len(df) // 10 for i in range(1, 10)]

# Divide el DataFrame en 100 partes
df_parts = np.split(df, split_points)

# Guarda cada parte como un archivo CSV separado
for i, df_part in enumerate(df_parts, start=1):
    df_part.to_csv(f"part_{i}.csv", index=False, sep=";")
```
